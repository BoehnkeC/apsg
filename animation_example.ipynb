# Stereonet animations in APSG
``plane``, ``line``, ``vector`` and ``pole`` methods of ``StereoNet`` store created artists internally and could be easily animated by ``animate`` method.

```python
from apsg import *
# This is important for animation to work in notebook
plt.rcParams['animation.html'] = 'html5'
# just bit bigger than default
plt.rcParams['figure.figsize'] = (8, 6)
```

Firstly we create ``Group`` of plane to be animated

```python
g = G('90 45 270 45 0 45 180 45 45 90 135 90', Fol)
```

Than we simply sequentially plot, usng one of the above mention methods, individual steps of animation.

```python
s = StereoNet()
for t in np.arange(0, 360, 2):
    s.plane(g.rotate(Lin(90,0), t), color='r')
```

Finally we call ``animate`` method which returns ``matplotlib.animation.ArtistAnimation`` object which is automatically shown in Juyter notebooks.

```python
s.animate(interval=40)
```
